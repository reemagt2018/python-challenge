```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from datetime import datetime


# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "./cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90,15)
lng_range = (-180, 180,15)


```

# Generate Cities List

```python

# List for holding lat_lngs and cities
lat_lngs = []
cities = []
cities_api=[]
country_api=[]

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)
lat_lngs


```

```python
# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    country=citipy.nearest_city(lat_lng[0], lat_lng[1]).country_code

    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        cities_api.append(city.replace(' ','+'))
       

# Print the city count to confirm sufficient count
len(cities)
```

# Perform API CAlls

```python
# OpenWeatherMap API Key
api_key = api_keys.api_key

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 
query_url = url + "&q=" 
query_url
```

#  Data to retrieve for analysis
Temperature (F) vs. Latitude
Humidity (%) vs. Latitude
Cloudiness (%) vs. Latitude
Wind Speed (mph) vs. Latitude

```python

lat=[]
lon=[]
temp=[]
speed=[]
humidity=[]
cloudiness=[]
date1=[]
cityfound=[]
cnt = 0
setcnt= 1
total=len(cityfound)

print(f"Beginning Data Retrival")
print(f"------------------------")

for city in cities_api: 
    cnt = cnt + 1 
    response=requests.get(query_url + (city)).json()
    print (f"Processing record {cnt} of set {setcnt}  | {city}" )
    print (f"Retrieving weather information from {query_url + city}")
     
   # print (response)
    try:
        if 'coord' in response:
            cityfound.append(response['name'])
            lat.append(response['coord']['lat'])
            lon.append(response['coord']['lon'])
        if 'main' in response:
            temp.append(response['main']['temp_max'])
            humidity.append(response['main']['humidity'])
        if 'wind' in response:
            speed.append(response['wind']['speed'])
        if 'clouds' in response:
            cloudiness.append(response['clouds']['all'])
            
        if 'dt' in response:    
            date1.append(response['dt'])
        
        
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")           
    if (cnt % 50)==0:
        cnt=0
        setcnt=setcnt+1
        time.sleep(5) 
           
           
print("--------------------------")
print("Data Retrieval Complete")
print("--------------------------")
       

# print(f"The latitude information received is: {lat}")
# print(f"The temperature information received is: {temp}")
# print(f"The longitute information received is: {lon}")
# print(f"The speed information received is: {speed}")
# print(f"The humidity information received is: {humidity}")
# print(f"The cloudiness information received is: {cloudiness}")
# print(f"The city found information received is: {cityfound}")
# print(f"The date information received is: {date1}")

```

```python
print (len(cityfound))
print (len(lat))
print (len(temp))
print (len(speed))
print (len(humidity))
print (len(cloudiness))
```

```python
# create a data frame from cities, lat, temp, speed , humidity, cloudiness
weather_dict = {
    "city": cityfound,
    "lat": lat,
    "lon":lon,
    "temp": temp,
    "speed" : speed,
    "humidity": humidity,
    "cloudiness": cloudiness,
    "date1":date1
    
    
}
weather_data = pd.DataFrame(weather_dict)
weather_data.head() 

plt_date = datetime.fromtimestamp(weather_dict['date1'][0])\
                    .strftime('%Y-%m--%d %H:%M:%S').split()[0]
```

# Temperature vs Latitude 

```python
# Build a scatter plot for each data type

# Plot settings
def set_plot(x_title,x_limit,y_title,y_limit):
    plt.title(f"{y_title} vs {x_title} ({plt_date})")
    plt.ylabel(y_title)
    plt.xlabel(x_title)
    plt.grid(True)
    plt.xlim(x_limit)
    
    
 
    
# Temperature vs Latitude 
weather_data.plot(kind="scatter",x="lat",y="temp",grid=True,color="blue")

#plt.scatter(weather_data["lat"], weather_data["temp"], color="blue")

set_plot("Latitude",[-90,90],"Max Temperature (F)",[0,100])
plt.axvline(0, color='black',alpha=0.5)


# Save the figure
plt.savefig("Temperature vs Latitude.png")

# Show plot
plt.show()
```
