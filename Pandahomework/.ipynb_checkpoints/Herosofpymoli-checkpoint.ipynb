# Gender Demographics
 
Percentage and Count of Male Players
Percentage and Count of Female Players
Percentage and Count of Other / Non-Disclosed
 



```python
# Create the GroupBy object based on the "Gender"  and "SN" column
gender_group1 = purchase_data[["SN","Gender"]]
gender_group_df=gender_group1.drop_duplicates(subset=None, keep='first', inplace=False)
gender_group_df

total_uniqe_participants=gender_group_df["Gender"].count()
total_uniqe_participants

## Percentage and Count of Male Players
# Percentage and Count of Female Players
# Percentage and Count of Other / Non-Disclosed
gender_data_final=gender_group_df["Gender"].value_counts()
df_gender_final=pd.DataFrame(gender_data_final)
df_gender_final.reset_index(inplace=True)
df_gender_final

```

```python
df_gender_final["Percentage"] = (df_gender_final["Gender"]/total_uniqe_participants) * 100

df_gender_final["Percentage"]=df_gender_final["Percentage"].map("{:.2f}".format)
    
df_gender_final = df_gender_final.rename(columns={"index":"Gender","Gender":"Total Count", "Percentage":"Percentage of Players"})
df_gender_final
```

```python
df_gender_final_tmp=df_gender_final.set_index("Gender")
df_gender_final_tmp
```

#  Purchasing Analysis (Gender)
 
Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
Create a summary data frame to hold the result
Optional: give the displayed data cleaner formatting
Display the summary data frame

```python
def f(x):
    d = {}
    d['purchase_cnt'] = x['Purchase ID'].count()
    d['avg_purchase_price'] = x['Price'].mean()
    d['total_purchase_price'] = x['Price'].sum()
    
    return pd.Series(d, index=['purchase_cnt', 'avg_purchase_price','total_purchase_price'])
    

purchasing_df=purchase_data.groupby(["Gender"]).apply(f)
purchasing_df["avg_purchase_total_per_person"]=purchasing_df["total_purchase_price"]/df_gender_final_tmp["Total Count"]

# Format the numbers
purchasing_df["avg_purchase_price"]=purchasing_df["avg_purchase_price"].map("${:.2f}".format)

purchasing_df["avg_purchase_total_per_person"]=purchasing_df["avg_purchase_total_per_person"].map("${:.2f}".format)
purchasing_df["total_purchase_price"]=purchasing_df["total_purchase_price"].map("${:.2f}".format)
purchasing_df


```

# Age Demographics
 
Establish bins for ages
Categorize the existing players using the age bins. Hint: use pd.cut()
Calculate the numbers and percentages by age group
Create a summary data frame to hold the results
Optional: round the percentage column to two decimal points
Display Age Demographics Table


```python
purchase_data['bin'] = pd.cut(purchase_data['Age'], [0,9, 14, 19,24,29,34,39,200] ,
                              labels=['<10', '10-14', '15-19','20-24','25-29','30-34','35-39',' 40+'])

purchase_data.head()
```

```python
bin_group1 = purchase_data[["SN","bin"]]
bin_group_df=bin_group1.drop_duplicates(subset=None, keep='first', inplace=False)
bin_group_df

total_participant_bin=bin_group_df["bin"].count()
total_participant_bin

## Percentage and Count of Male Players
# Percentage and Count of Female Players
# Percentage and Count of Other / Non-Disclosed

bin_data_final=bin_group_df["bin"].value_counts()
df_bin_final=pd.DataFrame(bin_data_final)
df_bin_final
total_participant_bin

```

```python
df_bin_final["Percentage"] = (df_bin_final["bin"]/total_participant_bin) * 100

df_bin_final["Percentage"]=df_bin_final["Percentage"].map("{:.2f}".format)
df_bin_final = df_bin_final.rename(columns={"bin":"Total Count", "Percentage":"Percentage of Players"})
df_bin_final.reset_index(inplace=True)
df_bin_final = df_bin_final.rename(columns={"index":"bin"})
df_bin_final=df_bin_final.set_index("bin")

df_bin_final

#df_bin_final.sort_index
```

#  Most Popular Items
 
Retrieve the Item ID, Item Name, and Item Price columns


Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


Create a summary data frame to hold the results


Sort the purchase count column in descending order


Optional: give the displayed data cleaner formatting


Display a preview of the summary data frame

```python
purchase_data.head()
popular_group1 = purchase_data[["Item ID","Item Name","Price"]]


def i(x):
    d = {}
    d['purchase_cnt'] = x['Item ID'].count()
    d['avg_purchase_price'] = x['Price'].mean()
    d['total_purchase_price'] = x['Price'].sum()
    return pd.Series(d, index=['purchase_cnt', 'avg_purchase_price','total_purchase_price'])

Item_df=popular_group1.groupby(["Item ID","Item Name"]).apply(i)


Item_df = Item_df.rename(columns={"purchase_cnt":"Purchase Count", "avg_purchase_price":"Item Price","total_purchase_price":"Total Purchase Value"})


Item_df_sorted = Item_df.sort_values(["Purchase Count"], ascending=False)
# Format the numbers
Item_df_sorted["Item Price"]=Item_df_sorted["Item Price"].map("${:.2f}".format)

Item_df_sorted["Total Purchase Value"]=Item_df_sorted["Total Purchase Value"].map("${:.2f}".format)



Item_df_sorted.head()


```

# Most Profitable Items
 
Sort the above table by total purchase value in descending order


Optional: give the displayed data cleaner formatting


Display a preview of the data frame

```python
Item_df_sorted_1 = Item_df_sorted.copy(deep=True)

Item_df_sorted_1["Total Purchase Value"] = Item_df_sorted['Total Purchase Value'].replace({'\$':''}, regex=True)

#Item_df_sorted_2 = pd.Item_df_sorted_1([1, 2], dtype='int32')
Item_df_sorted_1["Total Purchase Value"] = pd.to_numeric(Item_df_sorted_1["Total Purchase Value"])

Item_df_sorted_profit= Item_df_sorted_1.sort_values(["Total Purchase Value"], ascending=False)

Item_df_sorted_profit["Total Purchase Value"]=Item_df_sorted_profit["Total Purchase Value"].map("${:.2f}".format)
Item_df_sorted_profit.head()
```
